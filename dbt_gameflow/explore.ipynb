```python
import json
from pathlib import Path

import pandas as pd
import requests

F1_LEAGUE_ID = 4370
F1_SEASON = "2026"
API_KEY = 149076


def fetch_and_save_f1_season():
    url = (
        f"https://www.thesportsdb.com/api/v1/json/{API_KEY}/eventsseason.php"
        f"?id={F1_LEAGUE_ID}&s={F1_SEASON}"
    )

    response = requests.get(url)
    response.raise_for_status()
    data = response.json()

    events = data.get("events") or []

    if not events:
        print("No events found.")
        return

    # Keep only actual races
    races = [
        e for e in events
        if "grand prix" in e.get("strEvent", "").lower()
        and "qualifying" not in e.get("strEvent", "").lower()
        and "practice" not in e.get("strEvent", "").lower()
        and "sprint" not in e.get("strEvent", "").lower()
    ]

    if not races:
        print("No races found.")
        return

    df = pd.DataFrame(races)

    output_dir = Path("data")
    output_dir.mkdir(parents=True, exist_ok=True)
    output_file = output_dir / f"f1_season_{F1_SEASON}.parquet"

    df.to_parquet(output_file, index=False, engine="pyarrow")

    print(f"Saved F1 {F1_SEASON} season to {output_file}")
    print(f"Total races saved: {len(df)}")


if __name__ == "__main__":
    fetch_and_save_f1_season()
```
